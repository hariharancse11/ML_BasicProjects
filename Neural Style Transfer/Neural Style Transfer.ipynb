## Imports

```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

## Pretrained tensorflow model

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

## Preprocess image

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

## Load image

```python
target_image = load_image('frida.jpg')
style_image = load_image('starrynight.jfif')
```

## Visualize image

```python
target_image.shape
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
plt.imshow(np.squeeze(target_image))
plt.show()
```

## Stylize image

```python
stylized_image = model (tf.constant(target_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```
